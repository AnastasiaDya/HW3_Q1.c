```python
import pulp

def supplier_consumer_func(suppliers, consumers, edges):
#  הגדרת הבעיה
    problem = pulp.LpProblem("Supplier_Consumer_LP", pulp.LpMaximize)

# (Xij) יצירת משתנה לכל צלע
    x = {
        (u, v): pulp.LpVariable(f"x_{u}_{v}", lowBound=0)
        for (u, v) in edges
    }
# יצירת פונקציית המטרה
    problem += pulp.lpSum(x[(u, v)] for (u, v) in edges)

#אילוצים
    for u in suppliers:
        problem += pulp.lpSum(x[(i, j)] for (i, j) in edges if i == u) <= suppliers[u]

    for v in consumers:
        problem += pulp.lpSum(x[(i, j)] for (i, j) in edges if j == v) <= consumers[v]

    problem.solve(pulp.PULP_CBC_CMD(msg=False))

#חילוץ הערכים
    optimal_value = pulp.value(problem.objective)

    flow = {
        (u, v): pulp.value(x[(u, v)])
        for (u, v) in edges
    }

    return optimal_value, flow


```

דוגמת הרצה

```python
suppliers = {
    "u1": 3.5,
    "u2": 2,
    "u3": 4,
    "u4": 1.5
}

consumers = {
    "v1": 2.5,
    "v2": 3,
    "v3": 2,
    "v4": 1.5,
    "v5": 2
}

edges = [
    ("u1", "v1"),
    ("u1", "v2"),
    ("u1", "v4"),
    ("u2", "v2"),
    ("u2", "v5"),
    ("u3", "v3"),
    ("u3", "v4"),
    ("u4", "v2")
]
```

```python
optimal_value, flow = supplier_consumer_func(suppliers, consumers, edges)

print("Optimal value:", optimal_value)
print("Flow on edges:")

for edge, value in flow.items():
    if value > 0:
        print(edge, ":", value)
```
